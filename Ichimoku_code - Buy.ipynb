```python
data_long_profits = pd.DataFrame({
        "Open": data_signals.loc[(data_signals["Signal"] == "Buy") &
                                  data_signals["Regime"] == 1, "Open"],
        "High":data_signals.loc[(data_signals["Signal"] == "Buy") &
                                  data_signals["Regime"] == 1, "High"],
        "Low": data_signals.loc[(data_signals["Signal"] == "Buy") &
                                  data_signals["Regime"] == 1, "Low"],
        "Close": data_signals.loc[(data_signals["Signal"] == "Buy") &
                                  data_signals["Regime"] == 1, "Close"],
        "End Date": data_signals["Open"].loc[
            data_signals.loc[(data_signals["Signal"].shift(1) == "Buy") & (data_signals["Regime"].shift(1) == 1)].index
        ].index,
        "Pft-Price": data_signals.loc[(data_signals["Signal"].shift(1) == "Buy") & (data_signals["Regime"].shift(1) == 1),'Open'].values,
        "Profit": pd.Series(data_signals["Open"] - data_signals["Open"].shift(1)).loc[
            data_signals.loc[(data_signals["Signal"].shift(1) == "Buy") & (data_signals["Regime"].shift(1) == 1)].index
        ].tolist()
})
data_long_profits
```

```python
tradeperiods = pd.DataFrame({"Start": data_["Close"].loc[data_.loc[(data_["Signal"].shift(1) == 1) & (data_["Regime"].shift(1) == 1)].index].index ,
                            "End": data_long_profits["End Date"]})
data_long_profits["Trade Period Low"] = tradeperiods.apply(lambda x: min(data_.loc[x["Start"]:x["End"], "Low"]), axis = 1)
data_long_profits
```

```python
cash = 1000000
data_backtest = pd.DataFrame({"Start Port. Value": [],
                         "End Port. Value": [],
                         "End Date": [],
                         "Shares": [],
                         "Share Price": [],
                         "Trade Value": [],
                         "Profit per Share": [],
                         "Total Profit": [],
                         "Stop-Loss Triggered": []})
port_value = .1  # Max proportion of portfolio bet on any trade
batch = 100      # Number of shares bought per batch
stoploss = .02    # % of trade loss that would trigger a stoploss
for index, row in data_long_profits.iterrows():
    batches = np.floor(cash * port_value) // np.ceil(batch * row["Open"]) # Maximum number of batches of stocks invested in
    trade_val = batches * batch * row["Open"] # How much money is put on the line with each trade
    if row["Trade Period Low"] < (1 - stoploss) * row["Open"]:   # Account for the stop-loss
        share_profit = row["Trade Period Low"] - row["Open"]
        sqr_pft = row["Trade Period Low"]
        stop_trig = True
    else:
        share_profit = row['Pft-Price'] - row["Open"]
        sqr_pft = row["Pft-Price"]
        stop_trig = False
    profit = share_profit * batches * batch # Compute profits
    # Add a row to the backtest data frame containing the results of the trade
    data_backtest = data_backtest.append(pd.DataFrame({
                "Start Port. Value": cash,
                "End Port. Value": cash + profit,
                "End Date": row["End Date"],
                "Shares": batch * batches,
                "Share Price": row["Open"],
                "Trade Value": trade_val,
                "Profit per Share": share_profit,
                "Square price": sqr_pft,
                "Total Profit": profit,
                "Stop-Loss Triggered": stop_trig
            }, index = [index]))
    cash = max(0, cash + profit)
 
data_backtest
```
